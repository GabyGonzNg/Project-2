## 1. Function Q: Water Weight (lb/yd³)
▶️ **Code Cell:** define `calc_water_weight(...)`

**Inputs:**
- cement content
- w/c ratio
- (<any modifiers from Excel>)

**Output:**
- water weight (lb/yd³)

---


```python
def calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, wc_ratio_E):
    total_cementitious = cement_A + fly_ash_B + silica_fume_C + other_scm_D
    water_weight_Q = total_cementitious * wc_ratio_E
    return water_weight_Q
```

## 2. Function R: Cement Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_R(...)`

**Inputs:**
- weight of cement A
- Specific gravity J
- Unit Weight of water
  
**Output:**
- Cement Volume R (ft³)

---


```python
def calculate_volume_R(cement_A, sg_cement_J):
    volume_R = cement_A / (sg_cement_J * unit_weight_water)
    return volume_R
```

## 3. Function S: Fly Ash Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_S(...)`

**Inputs:**
- weight of Fly Ash B
- Specific gravity K
- Unit Weight of water
  
**Output:**
- Cement Volume S (ft³)

---


```python
def calculate_volume_S(fly_ash_B, sg_fly_ash_K):
    volume_S = fly_ash_B / (sg_fly_ash_K * unit_weight_water)
    return volume_S
```

## 4. Function T: Silica Fume Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_T(...)`

**Inputs:**
- weight of Silica Fume C
- Specific gravity L
- Unit Weight of water
  
**Output:**
- Cement Volume T (ft³)

---


```python
def calculate_volume_T(silica_fume_C, sg_silica_fume_L):
    volume_T = silica_fume_C / (sg_silica_fume_L * unit_weight_water)
    return volume_T
```

## 5. Function U: Other SCM Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_R(...)`

**Inputs:**
- weight of other SCM D
- Specific gravity M
- Unit Weight of water
  
**Output:**
- Cement Volume U (ft³)

---


```python
def calculate_volume_U(other_scm_D, sg_other_scm_M):
    return other_scm_D / (sg_other_scm_M * unit_weight_water)
```

## 6. Function V: Air Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_V(...)`

**Inputs:**
- Target air content F
- total volume of 1 cubic yard
  
**Output:**
- Air Volume V (ft³)

---

```python
def calculate_air_volume_V(target_air_content_F):
    volume_V = (target_air_content_F/100) * cubic_yard_ft3
    return volume_V
```

## 7. Function W: Water Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_W(...)`

**Inputs:**
- Weight of water Q
- unit weight of water
  
**Output:**
- Water Volume W (ft³)

---

```python
def calculate_water_volume_W(water_weight_Q):
    volume_W = water_weight_Q / unit_weight_water
    return volume_W
```

## 8. Function X: Total Aggregate Volume (ft³)
▶️ **Code Cell:** define `calculate_volume_W(...)`

**Inputs:**
- total volume of 1 cubic yard
- Cemetitious material volumes (R,S,T,U)
- air volume V
- water volume W
  
**Output:**
- Total Aggregate Volume (ft³)

---

```python
def calculate_total_aggregate_volume_X(R_cement, S_cement, T_cement, U_cement, V_air_volume, W_water_volume):
    volume_x = cubic_yard_ft3 - R_cement - S_cement - T_cement - U_cement - V_air_volume - W_water_volume
    return volume_x
```

## 9. Function Y: Fine Aggregate Weight (lb)
▶️ **Code Cell:** define `calculate_fine_aggregate_weight_Y(...)`

**Inputs:**
- target percent fine aggregate G
- specific gravity of fine aggregate N
- total aggregate volume X
- unit weight of water
  
**Output:**
- Fine Aggregate Weight (lb)

---

```python
def calculate_fine_aggregate_weight_Y(G_percent_fine_aggregate, N_specific_gravity, X_total_aggregate_volume):
    volume_y = unit_weight_water * (G_percent_fine_aggregate/100) * N_specific_gravity * X_total_aggregate_volume
    return volume_y
```

## 10. Function Z: Coarse Aggregate Weight (lb)
▶️ **Code Cell:** define `calculate_coarse_aggregate_weight_Z(...)`

**Inputs:**
- target percent coarse aggregate H
- specific gravity of coarse aggregate O
- total aggregate volume X
- unit weight of water
  
**Output:**
- Coarse Aggregate Weight (lb)

---

```python
def calculate_coarse_aggregate_weight_Z(H_percent_coarse_aggregate, O_specific_gravity, X_total_aggregate_volume):
    volume_z = unit_weight_water * (H_percent_coarse_aggregate/100) * O_specific_gravity * X_total_aggregate_volume
    return volume_z
```

## 11. Function AA: Other Aggregate Weight (lb)
▶️ **Code Cell:** define `calculate_other_aggregate_weight_AA(...)`

**Inputs:**
- target percent other aggregate I
- specific gravity of other aggregate P
- total aggregate volume X
- unit weight of water
  
**Output:**
- Other Aggregate Weight (lb)

---

```python
def calculate_other_aggregate_weight_AA(I_percent_other_aggregate, P_specific_gravity, X_total_aggregate_volume):
    volume_AA = unit_weight_water * (I_percent_other_aggregate/100) * P_specific_gravity * X_total_aggregate_volume
    return volume_AA
```

# Part 4 - Running Complete Mix Design

Run your functions, output a table of the design (Lab #3 will show you how to do that)

```python
# NDOT Concrete Mix Design – Full Integrated Run

def ndot_mix_design():

    print("\n=== NDOT Concrete Mix Design ===")
    # Project metadata
    project_no = int(input("Enter project number: "))
    control_no = int(input("Enter control number: "))
    concrete_class = input("Enter class of concrete: ")

    # Cementitious material inputs (lb per cubic yard)
    cement_A = float(input("Enter weight of cement per yd^3: "))
    fly_ash_B = float(input("Enter weight of fly ash per yd^3: "))
    silica_fume_C = float(input("Enter weight of silica fume per yd^3: "))
    other_scm_D = float(input("Enter weight of other SCM per yd^3: "))

    wc_ratio_E = float(input("Enter targer water/cement ratio: "))
    target_air_content_F = float(input("Enter targer % air content: "))


    # Aggregate proportions (%)
    G_percent_fine_aggregate = float(input("Enter targer % fine aggregate: "))
    H_percent_coarse_aggregate = float(input("Enter targer % coarse aggregate: "))
    I_percent_other_aggregate = float(input("Enter targer % other aggregate: "))

    # Specific gravities
    sg_cement_J = float(input("Enter specific gravity of cement J: "))
    sg_fly_ash_K = float(input("Enter specific gravity of cement K: "))
    sg_silica_fume_L = float(input("Enter specific gravity of cement L: "))
    sg_other_scm_M = float(input("Enter specific gravity of cement M: "))
    N_specific_gravity = float(input("Enter specific gravity of cement N: "))
    O_specific_gravity = float(input("Enter specific gravity of cement O: "))
    P_specific_gravity = float(input("Enter specific gravity of cement P: "))

    # Calculations
    Q = calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, wc_ratio_E)

    R = calculate_volume_R(cement_A, sg_cement_J)
    S = calculate_volume_S(fly_ash_B, sg_fly_ash_K)
    T = calculate_volume_T(silica_fume_C, sg_silica_fume_L)
    U = calculate_volume_U(other_scm_D, sg_other_scm_M)

    V = calculate_air_volume_V(target_air_content_F)
    W = calculate_water_volume_W(Q)

    X = calculate_total_aggregate_volume_X(R, S, T, U, V, W)

    Y = calculate_fine_aggregate_weight_Y(G_percent_fine_aggregate, N_specific_gravity, X)
    Z = calculate_coarse_aggregate_weight_Z(H_percent_coarse_aggregate, O_specific_gravity, X)
    AA =  calculate_other_aggregate_weight_AA(I_percent_other_aggregate, P_specific_gravity, X)

    # Output: Final Weight Chart
    print("\n---------------------------------------------")
    print(" NDOT Concrete Mix Design – Weight Summary")
    print("         (1 Cubic Yard of Concrete)")

    print("---------------------------------------------")
    print(f"Project Number:       {project_no}")
    print(f"Class of Concrete:    {concrete_class}")
    print(f"Control Number:       {control_no}")
    print("---------------------------------------------")
    print(f"Cement (A):            {cement_A:8.1f} lb")
    print(f"Fly Ash (B):           {fly_ash_B:8.1f} lb")
    print(f"Silica Fume (C):       {silica_fume_C:8.1f} lb")
    print(f"Other SCM (D):         {other_scm_D:8.1f} lb")
    print("---------------------------------------------")
    print(f"Fine Aggregate (Y):    {Y:8.0f} lb")
    print(f"Coarse Aggregate (Z):  {Z:8.0f} lb")
    print(f"Other Aggregate (AA):  {AA:8.0f} lb")
    print("---------------------------------------------")
    print(f"Water (Q):             {Q:8.0f} lb")
    print("---------------------------------------------")
    print("End of Mix Design Summary")

    # ---- Prepare NDOT-style table ----
    components = [
        "Cement",
        "Water",
        "Fine Aggregate",
        "Coarse Aggregate",
        "Other Aggregate",
        "Air (volumetric)",
        "Total (sanity check)"
    ]

    weights = [
        round(cement_A, 1),
        round(Q, 1),
        round(Y, 1),
        round(Z, 1),
        round(AA, 1),
        f"{round(target_air_content_F, 1)} %",  # volumetric air
        round(cement_A + Q + Y + Z + AA, 1)
    ]

    notes = [
        "",  # Cement
        "",  # Water
        "",  # Fine Aggregate
        "",  # Coarse Aggregate
        "",  # Other Aggregate
        "percent / ft³",  # Air
        ""  # Total
    ]

    df_ndot = pd.DataFrame({
        "Component": components,
        "Weight (lb/yd³)": weights,
        "Notes": notes
    })

    # ---- Display NDOT-style table ----
    print(f"\nNDOT Concrete Mix Design – Full Summary (1 yd³ of Concrete)")
    print(f"Project Number: {project_no} | Class: {concrete_class} | Control No: {control_no}")
    print("-------------------------------------------------------------")
    display(df_ndot)
```
